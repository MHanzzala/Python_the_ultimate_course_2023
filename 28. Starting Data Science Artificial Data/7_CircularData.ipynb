```python
import matplotlib.pyplot as plt
import numpy as np

def addOutlier(l, outlier):
    l=list(l)
    for item in outlier:
        l.append(item)
    return l

rng = np.random.RandomState()


x = 10*rng.rand(50)
y = 2*x*x*x -x*x+x +  rng.rand(50)
outlierX=[0.2,4,8, 9]
outlierY=[5,2.5,17.5, 2.7]

xNew= addOutlier(x,outlierX)
yNew= addOutlier(y,outlierY)


plt.scatter(x,y)
plt.scatter(outlierX,outlierY, color="red")
```
